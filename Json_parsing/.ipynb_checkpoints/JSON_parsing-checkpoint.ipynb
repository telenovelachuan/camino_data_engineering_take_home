```python
import json
import pprint
import uuid
import xmltodict

RESULT_TABLE_NAMES = ['prequal_results', 'xml_details', 'products', 'credit_profiles', 'headers', 'risk_models', 'consumer_identities']
with open("./data/prequalresult.json", 'r') as f:
    json_file = json.load(f)
json_file
```

Building PrequalResult table objects
